The issue I am solving here is that the processed_churn_data file did not include the customerIDs (because it was made partly using synthetic data). To combat this, I removed the synthetic data and appended the customerIDs inside -> Raw_Data.csv

After which I performed prediction using LR model (as it had highest accuracy) on the TEST dataset (as predictions are often made on non-trained data) and stored the predicted results in Prediction_Results.csv

```python
import pandas as pd

file_path = 'processed_churn_data.csv'
df = pd.read_csv(file_path)

subset_data = df.iloc[:7032]
```

```python
subset_data
```

```python
customer_data_path = 'TelcoCustomerChurn.csv'
customer_data = pd.read_csv(customer_data_path)

customer_id_column = customer_data['customerID']

# Append the CustomerID column to the subset data
subset_data_with_id = subset_data.copy()
subset_data_with_id['customerID'] = customer_id_column[:7033].values

```

```python
subset_data_with_id
```

```python
# Append a placeholder to the CustomerID column to match the length
if len(customer_id_column) < len(subset_data):
    customer_id_column = customer_id_column.append(pd.Series([None] * (len(subset_data) - len(customer_id_column))), ignore_index=True)

# Add the CustomerID column to the subset data
subset_data_with_id = subset_data.copy()
subset_data_with_id['customerID'] = customer_id_column[:7033].values

# Move the CustomerID column to the first position
cols = subset_data_with_id.columns.tolist()
cols = cols[-1:] + cols[:-1]
subset_data_with_id = subset_data_with_id[cols]

```

```python
subset_data_with_id
```

```python
subset_data_with_id.to_csv("Raw_Data.csv",index=False)
```

```python
import pandas as pd
import pickle

data_path = 'Raw_Data.csv'
data = pd.read_csv(data_path)

model_path = 'PickleFiles/gb_model.pkl'  
with open(model_path, 'rb') as file:
    model = pickle.load(file)


X = data.drop(columns=['customerID', 'Churn'], errors='ignore')

predictions = model.predict(X)

results = pd.DataFrame({
    'customerID': data['customerID'],
    'Churn': predictions
})

# Save the results to a new CSV file
output_path = 'Prediction_Results.csv'
results.to_csv(output_path, index=False)

output_path
```

```python
import pandas as pd

df = pd.read_csv('Raw_Data.csv')

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop(['Churn', 'customerID'], axis=1)
y = df['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(max_iter=10000, solver="liblinear")
log_reg.fit(X_train, y_train)
```

```python
y_pred = log_reg.predict(X_test)
```

```python
y_pred
```

```python
results = pd.DataFrame({
    'customerID': df['customerID'].iloc[y_test.index],
    'Churn': y_pred
})


# Save the results to a CSV file
output_path = 'Prediction_Results.csv'
results.to_csv(output_path, index=False)
```
